### Open times are exponentially, not normally distributed, so we cannot use a t-Test.

### Use a permutation test to decide whether or not each data set came from the same distribution. i.e. whether or not the mutation had any effect on open time?

```python
# MLE time constants (tau) from each data set
loc, mleWildtypeTau_sec = st.expon.fit(wildtypeOpenTimes_sec, floc=0)
loc, mleMutantTau_sec = st.expon.fit(mutantOpenTimes_sec, floc=0)

# difference bewteen wild type and mutant taus
measuredTauDiff = mleWildtypeTau_sec - mleMutantTau_sec

# concatenate all open times from both wild type and mutant into one big array
nA = len(wildtypeOpenTimes_sec)
nB = len(mutantOpenTimes_sec)
allOpenTimes = np.zeros((nA + nB,))
allOpenTimes[:nA] = wildtypeOpenTimes_sec
allOpenTimes[-nB:] = mutantOpenTimes_sec

# generate 1000 permutations of mixtures of open times from both wild type and mutant
# for each permutation store the difference between the taus for the permuted data sets
numPermutations = 1000
permutedTauDiffs = np.zeros((numPermutations,))

for i in range(numPermutations):
    np.random.shuffle(allOpenTimes)
    
    permutedWildtypeOpenTimes = allOpenTimes[:nA]
    permutedMutantOpenTimes = allOpenTimes[-nB:]
    
    loc, permutedWildtypeTau = st.expon.fit(permutedWildtypeOpenTimes, floc=0)
    loc, permutedMutantTau = st.expon.fit(permutedMutantOpenTimes, floc=0)
    
    permutedTauDiffs[i] = permutedWildtypeTau - permutedMutantTau

# plot the distribution differences between taus for each permutation
plt.hist(permutedTauDiffs, bins=50, alpha=0.5, label='expected if H0 true');
# add 95% confidence intervals to the plot
lb, ub = np.quantile(permutedTauDiffs, [0.025, 0.975])
plt.axvline(lb, linestyle='--')
plt.axvline(ub, linestyle='--');
# add the measured difference from the MLE fits
plt.axvline(measuredTauDiff, color='r', label='measured')
plt.xlabel('Wild type - Mutant Tau (s)')
plt.ylabel('# Permutations')
plt.legend();
```

### Compute a p-value for the observed difference in time constant between wild type and mutant to be explained by the null hypothesis that the mutation has no effect on open time.

```python
numPermutationsMoreExtreme = (np.abs(permutedTauDiffs) > np.abs(measuredTauDiff)).sum()
numPermutationsMoreExtreme, numPermutations
```

```python
pvalue = numPermutationsMoreExtreme / numPermutations
pvalue
```

### Reject the null hypothesis ==> mutation did affect open time.

### Compute effect size.

### For Cohen's d we need to divide the measured change in time constant between mutant and wild type by the standard deviation of the time constant distribution for either wild type or mutant.

### But we only have a single measurement for the wild type and mutant time constant.

### Ok, let's bootstrap to get the estimated standard deviation in the wild type time constant.

```python
wtTaus_sec = np.zeros((1000,))
for i in range(1000):
    bs_opentimes_sec = np.random.choice(wildtypeOpenTimes_sec, 
                                        size=wildtypeOpenTimes_sec.shape, replace=True)
    loc, wtTaus_sec[i] = st.expon.fit(bs_opentimes_sec, floc=0)

plt.hist(wtTaus_sec, bins=30, density=True, alpha=0.5, label='Bootstrapped')

# overlay MLE fit
mu, sd = st.norm.fit(wtTaus_sec)
taus = np.linspace(0.7, 1.05, 100)
plt.plot(taus, st.norm.pdf(taus, mu, sd), label='MLE')

plt.xlabel('Wild Type Taus (s)')
plt.ylabel('pdf')
plt.legend();

sd
```

```python
effectSize = measuredTauDiff / sd
effectSize
```

### Effect size is large. Mutation has a big effect on open time.
